```python
from matplotlib import pyplot as plt
import scipy.io as sio
import numpy as np
```

```python
weight_file_name = '180712_FCAIDE_FT_384000trdata_patch50_std30QED_x22_x64_residual_block_x1_output_x128_1by1_blocks_x2_layer_Result.mat'
f = sio.loadmat('./'+weight_file_name)
loss_arr_epoch = np.array(f["loss_arr_epoch"])
SSIM_arr_epoch = np.array(f["SSIM_arr_epoch"])
PSNR_arr_epoch = np.array(f["PSNR_arr_epoch"])
sup_PSNR_arr = np.array(f["sup_PSNR_arr"])
sup_SSIM_arr = np.array(f["sup_SSIM_arr"])
max_PSNR_denoised_image_arr = np.array(f["max_PSNR_denoised_image_arr"])
max_SSIM_denoised_image_arr = np.array(f["max_SSIM_denoised_image_arr"])
```

```python
print (loss_arr_epoch.shape)
print (SSIM_arr_epoch.shape)
print (PSNR_arr_epoch.shape)
print (sup_PSNR_arr.shape)
print (sup_SSIM_arr.shape)
print (max_PSNR_denoised_image_arr.shape)
print (max_SSIM_denoised_image_arr.shape)

ep_size = PSNR_arr_epoch.shape[1]
```

```python
plt.figure(figsize=[10,8])
plt.plot(np.mean(PSNR_arr_epoch,axis=0), linewidth=2, label='Sigma=30')
# plt.ylim([28.4,29.25])
plt.xlim([0,ep_size])
plt.xlabel('Epoch',fontsize=20)
plt.ylabel('PSNR',fontsize=20)
plt.legend(loc='best', fontsize=12)
#plt.savefig('./result_data/'+self.save_file_name+'_trloss.png')
plt.show()
```

```python
plt.figure(figsize=[10,8])
plt.plot(np.mean(loss_arr_epoch,axis=0), linewidth=2, label='Sigma=30')
# plt.ylim([28.4,29.25])
plt.xlim([0,ep_size])
plt.xlabel('Epoch',fontsize=20)
plt.ylabel('Loss',fontsize=20)
plt.legend(loc='best', fontsize=12)
#plt.savefig('./result_data/'+self.save_file_name+'_trloss.png')
plt.show()
```

```python
print ('average sup PSNR : '+str(np.mean(sup_PSNR_arr)))
print ('np.max(PSNR_arr_epoch) : ' + str(np.max(np.mean(PSNR_arr_epoch,axis=0))))
print ('np.argmax(PSNR_arr_epoch) : ' + str(np.argmax(np.mean(PSNR_arr_epoch,axis=0))))
```

```python
print ('average sup SSIM : '+str(np.mean(sup_SSIM_arr)))
print ('np.max(SSIM_arr_epoch) : ' + str(np.max(np.mean(SSIM_arr_epoch,axis=0))))
print ('np.argmax(SSIM_arr_epoch) : ' + str(np.argmax(np.mean(SSIM_arr_epoch,axis=0))))
```
