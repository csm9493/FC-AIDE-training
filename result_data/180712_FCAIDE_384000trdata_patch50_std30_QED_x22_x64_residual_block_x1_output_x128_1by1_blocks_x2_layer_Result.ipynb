```python
from matplotlib import pyplot as plt
import scipy.io as sio
import numpy as np
```

```python
weight_file_name = '180712_FCAIDE_384000trdata_patch50_std30_QED_x22_x64_residual_block_x1_output_x128_1by1_blocks_x2_layer_Result.mat'
f = sio.loadmat('./'+weight_file_name)
PSNR_arr_epoch = np.array(f["PSNR_arr_epoch"]).transpose()
loss_arr_epoch = np.array(f["loss_arr_epoch"]).transpose()
max_PSNR_denoised_images = np.array(f["max_PSNR_denoised_images"])
```

```python
print (PSNR_arr_epoch.shape)
print (loss_arr_epoch.shape)
print (max_PSNR_denoised_images.shape)

ep_size = PSNR_arr_epoch.shape[0]
```

```python
plt.figure(figsize=[10,8])
plt.plot(PSNR_arr_epoch, linewidth=2, label='Sigma=30')
# plt.ylim([28.4,29.25])
plt.xlim([0,ep_size])
plt.xlabel('Epoch',fontsize=20)
plt.ylabel('PSNR',fontsize=20)
plt.legend(loc='best', fontsize=12)
#plt.savefig('./result_data/'+self.save_file_name+'_trloss.png')
plt.show()
```

```python
plt.figure(figsize=[10,8])
plt.plot(loss_arr_epoch, linewidth=2, label='Sigma=30')
# plt.ylim([28.4,29.25])
plt.xlim([0,ep_size])
plt.xlabel('Epoch',fontsize=20)
plt.ylabel('Loss',fontsize=20)
plt.legend(loc='best', fontsize=12)
#plt.savefig('./result_data/'+self.save_file_name+'_trloss.png')
plt.show()
```
